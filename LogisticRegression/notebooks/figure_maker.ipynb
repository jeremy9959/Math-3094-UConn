```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.plotting import row
from bokeh.io import export_png
output_notebook()
import numpy as np
```

```python
xs=[str(x) for x in [-3,-2,-1,0,1,2,3]]
occur = [10,18,38,50,69,78,86]
p=figure(x_range=xs)
p.vbar(x=xs,top=[.1,.18,.38,.5,.69,.78,.86],width=.8)
show(p)
```

```python
xpts=[]
ypts=[]
for x,z in zip([-3,-2,-1,0,1,2,3],occur):
    xpts = xpts + [x]*100
    ypts = ypts + [1]*z+[0]*(100-z)
q=figure()
q.scatter(x=[x+np.random.normal(0,.1) for x in xpts],y=ypts,alpha=.1)
show(q)
```

```python
H=row([p,q])
export_png(H,filename="logistic_plots.png")
```
