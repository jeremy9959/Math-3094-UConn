```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, Y = np.meshgrid(np.linspace(-2,2,100),np.linspace(-2,2,100))
```

```python
Z=np.exp(-(X**2-X*Y+Y**2)/2)
```

```python
Z.shape
```

```python
X.shape
```

```python
fig,axes=plt.subplots(1)
axes.set_title('Level Curves of Multivariate Gaussian')
axes.contour(X,Y,Z,10)
```

```python
from scipy.stats import norm
```

```python
f=norm(0,1).rvs(1000)
```

```python
g = norm(0,1).rvs(1000)
```

```python
z = np.sqrt(f**2+g**2)
```

```python
k=plt.hist(z,bins=100,density=True)
```

```python
plt.plot(k[1][1:],np.cumsum(k[0])/np.sum(k[0]))
plt.plot(k[1][1:],1-np.exp(-k[1][1:]**2/2))
```

```python
x=np.linspace(0,1,100)
y=np.exp(55*np.log(x)+45*np.log(1-x))
y=100*y/np.sum(y)
```

```python
f=plt.figure()
f.set_size_inches=(10,10)
ax = f.add_subplot(1,1,1)
ax.plot(x,y)
ax.set_xticks(np.arange(0,1.1,.1))
```
