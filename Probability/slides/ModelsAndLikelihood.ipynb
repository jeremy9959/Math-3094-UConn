```python
import numpy as np
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
output_notebook()
```

```python
x=np.linspace(0,1,100)
y = 55*np.log(x)+45*np.log(1-x)
```

```python
f=figure(title='Likelihood Function')
```

```python
f.line(x=x,y=y)
```

```python
show(f)
```
